# Introduction

__The double-slit experiment__ is a famous demonstration of quantum mechanics that shows how particles like electrons or photons can behave like waves. Here’s a simplified description:

__Setup__: Imagine a barrier with two narrow slits and a screen behind it. You send particles (like electrons or light photons) toward the barrier.

__Expectation__ (if particles behave like tiny bullets): You’d expect the particles to pass through one slit or the other, forming two distinct bands on the screen.

__Reality (wave-like behavior)__: Instead, when you don't measure which slit each particle goes through, they interfere like waves, creating a pattern of bright and dark bands on the screen—this is called an interference pattern.

__Observation__ (what happens when you measure which path the particles take): If you add a detector to determine which slit each particle passes through, the interference disappears, and you only get two bands—showing that observation affects the behavior.

This experiment highlights the fundamental weirdness of quantum mechanics: __particles can behave like waves, and measurement itself changes the outcome__. Pretty mind-bending, right?

# Entropy Change in the Double-Slit Experiment with Which-Path Measurement

Let's analyze how the entropy changes when we determine which slit an electron passes through in the double-slit experiment.

In the standard double-slit experiment without which-path information:
- The electron's state is a superposition: |ψ⟩ = (|L⟩ + |R⟩)/√2
  where |L⟩ is the electron's state when it passes thorugh left slit
  and |R⟩ is the electron's state when it passes thorugh the right slit.
- The density matrix is: ρ = |ψ⟩⟨ψ| = ½(|L⟩⟨L| + |L⟩⟨R| + |R⟩⟨L| + |R⟩⟨R|)

When we measure which slit the electron passes through:
- The state collapses to either |L⟩ or |R⟩ with 50% probability each
- The density matrix becomes: ρ' = ½(|L⟩⟨L| + |R⟩⟨R|)

## Entropy Calculation

The von Neumann entropy is given by:
S = -Tr(ρ ln ρ)

### Initial Entropy (no which-path info)
The initial state is pure (ρ² = ρ):
$S_{initial} = -Tr(ρ ln ρ) = 0$ (pure states have zero entropy)

Let's show that $S_{initial} = 0$ for pure states.

|L> and |R> are ortonormal states:
$|L> = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, |R> = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$

Then 

$|L⟩⟨L| = \begin{pmatrix} 1 \\ 0 \end{pmatrix} * \begin{pmatrix} 1 & 0 \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$

$|L⟩⟨R| = \begin{pmatrix} 1 \\ 0 \end{pmatrix} * \begin{pmatrix} 0 & 1 \end{pmatrix} = \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}$

$|R⟩⟨L| = \begin{pmatrix} 0 \\ 1 \end{pmatrix} * \begin{pmatrix} 1 & 0 \end{pmatrix} = \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix}$

$|R⟩⟨R| = \begin{pmatrix} 0 \\ 1 \end{pmatrix} * \begin{pmatrix} 0 & 1 \end{pmatrix} = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}$

=> ρ = |ψ⟩⟨ψ| = ½(|L⟩⟨L| + |L⟩⟨R| + |R⟩⟨L| + |R⟩⟨R|) = ½$\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix}$

Obviously, ρ² = ρ

ρ² =  1/4$\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} * \begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} = 1/4\begin{pmatrix} 2 & 2 \\ 2 & 2 \end{pmatrix} = ½\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix}$=ρ

__Let's verify now that S_initial = -Tr(ρ ln ρ) = 0__. 

Since ρ is a Hemitian matrix, it can be diagonalized in an orthonormal basis: $ρ = UDU^+$ where D is a diagonal matrix containing the eigenvalues $\lambda_{i}$ of ρ, and U is a unitary transformation.

We have $Tr(ρ ln ρ) = Tr(UDU^+ln(UDU^+)) = Tr(UDU^+Uln(D)U^+)$ and, since $U^+U=I$, $Tr(ρ ln ρ) = Tr(UDln(D)U^+).$
The Trace is invariant under cyclic permutations, which means $Tr(ρ ln ρ) = Tr(U^+UDln(D)) = Tr(DlnD).$

Thus, $Tr(ρ ln ρ) = \sum_{i}\lambda_{i}ln\lambda_{i}$ where we used the trace definition: sum of a diagonal elements of an array.

For all practical purposes, $S_{initial} = -Tr(ρ ln ρ) = - \sum_{i}\lambda_{i}ln\lambda_{i}$.


Let's find now the eigenvalues for ρ and respective entropy.

$det(ρ - \lambda*I) => (1/2 - \lambda)^2 -1/4 = 0 = (1/2 - \lambda -1/2)*(1/2 - \lambda + 1/2) = 0  = -\lambda*(-\lambda + 1) = 0 => \lambda_{1} = 0, \lambda_{2} = 1$

=> $S_{initial} = -(0ln0 + 1ln1) = -0ln0$, since ln1 = 0
Now, 0ln0 = $lim_{x->\infty}(1/xln(1/x))=lim_{x->\infty}(-ln(x)/x)=lim_{x->\infty}-(1/x)=0$. => __$S_{initial} = 0$__.

### Final Entropy (with which-path info)
The final state is a mixed state with probabilities (0.5, 0.5):
$S_{final} = -Σ p_{i} ln p_{i} = -[0.5 ln(0.5) + 0.5 ln(0.5)] = ln(2) ≈ 0.693 nats$
(or ≈ 1 bit in base 2)

$-[0.5ln(0.5)+0.5ln(0.5)]=-ln(0.5)=-ln(1/2)=ln(2)$

ρ' = ½(|L⟩⟨L| + |R⟩⟨R|) = ½$\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$ already diagonal with 1/2 eigenvalues.

=> When we measure which slit the electron passes through, L or R, we earn one bit of information (in logaritimic base of 2) corresponding to the above entropy increase.
The connection between entropy and quantum interference is indeed fundamental. In the double-slit experiment, the system starts in a pure quantum state with zero entropy. When we gain which-path information, the wavefunction collapses, and the system transitions to a mixed state with an entropy increase of ln(2). __This entropy increase reflects the loss of quantum coherence__—the interference disappears because we have acquired classical information.

In classical physics, entropy increase is tied to the arrow of time, meaning systems evolve irreversibly. But in the double-slit experiment, if we do not measure which-path information, the system remains in a pure superposition, where the quantum state evolves unitarily without increasing entropy. __This suggests that, from the perspective of quantum mechanics, there is no irreversible progression of time in the absence of measurement__. The system remains in an evolving quantum state that retains full coherence.

Even without measurement, the quantum state evolves deterministically according to the Schrödinger equation. What changes is that without decoherence (measurement), the system does not experience thermodynamic irreversibility—which is typically associated with the classical perception of time flowing forward. So, the relationship between entropy and time flow in quantum mechanics is subtle. In the double-slit experiment:

- Without measurement → Pure state, coherence retained, reversible evolution (unitary).

- With measurement → Mixed state, coherence lost, entropy increase, irreversible information gain.

This aligns with the idea that __irreversibility emerges when classical observers extract information from quantum systems__, reinforcing the deep connection between entropy, measurement, and our perception of time.

Thinking about entropy, decoherence and “which‐path” information in the double-slit leads naturally into quantum thermodynamics and an information-theoretic arrow of time:

- __Measurement, Decoherence and Entropy In the double-slit:__ You start in a pure superposition |ψ⟩=(|L⟩+|R⟩)/√2, ρ=|ψ⟩⟨ψ| ⇒ S=–Tr ρ ln ρ=0. Once you detect “left” vs. “right,” ρ→½(|L⟩⟨L|+|R⟩⟨R|), S= ln 2. This jump reflects loss of coherence (off-diagonal terms vanish) and emergence of irreversibility: you’ve imprinted path information into the detector and environment.

- __Landauer’s Principle & Thermodynamic Cost of Information:__ Landauer taught us that erasing (or, equivalently, acquiring and then forgetting) one bit of information must dissipate ≥k_BT ln 2 of heat into a bath, generating entropy there. In our slit case, gaining which-path info carries exactly that ln 2 of entropy away into the measurement apparatus and its surroundings.

- __Open Quantum Systems & the Arrow of Time:__ A truly isolated quantum system—like an unmeasured double-slit interferometer—evolves unitarily (via the Schrödinger equation) and conserves von Neumann entropy. Only when you couple it irreversibly to an environment (introducing decoherence described by a Lindblad master equation) does entropy in the system+environment increase and an arrow of time emerge.

- __Quantum Clocks, Entropy Generation & Timekeeping:__ Any physical clock must couple to something irreversible to record “ticks.” In quantum thermodynamics, even the best “quantum clock” leaks entropy into its environment with each tick. __Without that leakage there’d be no record of elapsed time—no way to distinguish “before” from “after.”__ __Thus entropy production isn’t just a side-effect; it’s fundamental to the operational meaning of time in quantum mechanics.__

- __Information-Theoretic Time and the Flow of Events:__ From an information perspective, “time flows” when new information about a system is irreversibly deposited into its environment. In the unmeasured double-slit, the quantum state changes—but no new classical information emerges, so there’s no thermodynamic “clock-hand” marking that change. Only when you gain which-path (or any other) information does the universe’s entropy budget tick upward and a true arrow of time appear.

# Many Worlds (Everett) view on Entropy

In the Many-Worlds (Everett) view there is no “mystical collapse”—only unitary evolution of the universal wavefunction. When you put a which-path detector in place, here’s what happens:

- __Before measurement:__ We have a superposition |Ψ⟩ = (|L⟩ + |R⟩)/√2 ⊗ |ready⟩ₙₒ where |L⟩,|R⟩ are the two slit paths and |ready⟩ₙₒ is “detector + environment in a neutral state.”
-  __Measuring: Detector and Environment entangle with the slit paths initianting branching:__ The detector interacts unitarily, entangling its pointer with the particle: |Ψ⟩ → (|L⟩⊗|D=L⟩⊗|E_L⟩  +  |R⟩⊗|D=R⟩⊗|E_R⟩) /√2. |D=L⟩,|D=R⟩ are “detector recorded left/right.” |E_L⟩,|E_R⟩ are distinct environment states (“air molecules, photons, memory bits, etc.”) that have picked up path-info via scattering, thermal radiation, brain neurons firing, and so on.
- __Imprint in all branches:__ There is no single “true” outcome—in each branch you (and your apparatus and environment) record a definite path. All branches coexist in the global wavefunction, but they no longer interfere because ⟨D=L,E_L | D=R,E_R⟩ ≈ 0 (decoherence kills off-diagonals, no overlap).
- __Entropy and Records:__ If you trace out (ignore) the environment+detector, the remaining particle state is mixed—Sₙᵥₙ = ln 2—because you’ve effectively “thrown away” which-branch distinctions. But the full state is still pure, just highly entangled.
- __Why interference vanishes:__ In principle you could recombine branches only if you perfectly reversed all recording processes (erasing every scattered photon, thermal kick, memory trace…). In practice this is impossible, so the branches behave like classical alternatives.

__Bottom line:__ Yes—under Many-Worlds every branch carries its own imprint of “which-path” in both detector and environment. You don’t collapse the wavefunction to one record; you only become correlated with one branch. Decoherence then makes those branches dynamically autonomous, giving you the appearance of a single recorded outcome.

Each branch of the universal wave-function carries its own local entropy—but the global wave-function remains pure (zero entropy). Here’s how to see it:

- Global purity vs. local mixedness – In Many-Worlds the total state Ψₜₒₜ = ∑ₖ cₖ |system: k⟩⊗|env: k⟩ is still a pure state ⇒ S [Ψₜₒₜ]=−Tr (ρₜₒₜ ln ρₜₒₜ)=0. – But any observer or subsystem looks at a reduced density matrix (trace out everything else). That reduced ρ generally is mixed, with S>0.
- Which-path branching and entropy – Before measurement: one global branch, pure ⇒ S=0. – After the which-path interaction you get two orthogonal branches: (|L⟩⊗|D=L⟩⊗|E_L⟩ + |R⟩⊗|D=R⟩⊗|E_R⟩)/√2 – If you “live” in the L‐branch and trace out R, D_R, E_R, you assign S=ln 2 to your reduced ρ. Likewise for the R-branch.
- Branch‐relative entropy vs. universal entropy – Branch entropy is a measure of your ignorance about the rest of the multiverse once you condition on one outcome. – Universal entropy stays zero because the wavefunction hasn’t collapsed — it’s just become highly entangled.
- Arrow of time in each world – Decoherence imprints records in the environment, pumping up local entropy and giving each branch its own arrow of time. – Globally, you could (in principle) unitarily rewind everything back to a pure superposition, but in practice each world “forgets” its off-diagonal phases forever.

Every decoherent branch locally picks up an entropy increase (e.g. ln 2 for which-path), but all those branches together remain in a single pure state of zero entropy. __Your experience of irreversibility and the arrow of time lives inside your branch’s nonzero entropy, not in the universal wavefunction.__

```python
## Python Calculation

#Here's code to verify this calculation:
import numpy as np
from scipy.linalg import logm

# Initial pure state density matrix
rho_initial = np.array([[0.5, 0.5], [0.5, 0.5]])

# Final mixed state density matrix
rho_final = np.array([[0.5, 0], [0, 0.5]])

def von_neumann_entropy(rho):
    eigenvalues = np.linalg.eigvalsh(rho)
    # Remove zero eigenvalues to avoid log(0)
    eigenvalues = eigenvalues[eigenvalues > 1e-10]
    print("eigenvalues:",eigenvalues)
    return -np.sum(eigenvalues * np.log(eigenvalues))

S_initial = von_neumann_entropy(rho_initial)
S_final = von_neumann_entropy(rho_final)

print(f"Initial entropy: {S_initial:.4f} nats")
print(f"Final entropy: {S_final:.4f} nats")
print(f"Entropy change: {S_final - S_initial:.4f} nats")
```

## Physical Interpretation

The entropy increases by ln(2) when we gain which-path information because:
1. We start with a pure quantum state (zero entropy)
2. The measurement process introduces classical uncertainty about the outcome
3. The system transitions from a quantum superposition to a classical statistical mixture

__This entropy increase is directly related to the destruction of quantum interference__ - the which-path measurement destroys the coherence between the |L⟩ and |R⟩ states, converting quantum uncertainty into classical uncertainty.

As you can see, there is some fundamental connection here between the quantum interference and entropy!

# Simulating a Double-Slit Experiment with Qiskit

We can model the key quantum effects of the double-slit experiment using qubits to represent:
- The "which slit" information (superposition)
- The interference pattern (phase manipulation)
- Measurement effects (collapse of the wavefunction)


## Key Features of This Simulation:

1. **Slit Representation**:
   - Qubit 0 represents which slit the particle passes through
   - In double-slit mode, it's put in superposition (Hadamard gate)
   - In single-slit mode, it remains |0⟩

2. **Interference Creation**:
   - Controlled rotations (CRY gates) create path-dependent phase shifts
   - CRY gates info: https://docs.quantum.ibm.com/api/qiskit/qiskit.circuit.library.CRYGate
   - Hadamard gates on screen qubits create interference
     

3. **Measurement Effects**:
   - When we measure which slit (qubit 0), superposition collapses
   - This destroys the interference pattern (decoherence)

4. **Screen Detection**:
   - The last 3 qubits represent different screen positions
   - The histogram shows the interference pattern

## Expected Results:

1. **Double-Slit (No Measurement)**:
   - Clear interference pattern (some positions more likely than others)

2. **Single-Slit Control**:
   - No interference (uniform distribution)

3. **Which-Path Measured**:
   - Interference disappears (resembles single-slit case)

We explicitly track and display which-slit information alongside the screen detection pattern. This will help visualize how measuring the slit affects the interference.

This simulation captures the key quantum phenomena of the double-slit experiment while working within the framework of gate-based quantum computing. The interference pattern emerges from the quantum superposition and phase relationships between paths.

## Details:

1. **Explicit Slit Measurement Tracking**:
   - Dedicated classical bit to record slit measurement (bit 0)
   - Separately counts slit measurement outcomes ('0' for left, '1' for right)

2. **Three-Panel Visualization**:
   - **Left Panel**: Shows the interference pattern at the detection screen
   - **Middle Panel**: Displays which-slit measurement results when taken
   - **Right Panel**: Shows example full measurement outcomes (slit + screen bits)

3. **Data Processing**:
   - Returns three separate count dictionaries:
     - `full_counts`: Complete measurement records
     - `screen_counts`: Just the screen detection pattern
     - `slit_counts`: Which-slit measurement results

4. **Measurement Control**:
   - The `measure_slit` parameter explicitly controls whether we:
     - Just measure the screen (preserves interference)
     - Also measure which slit (destroys interference)

## How to Interpret the Results when measuring which slit:
   - Middle panel shows ~50/50 distribution between slits (quantum superposition collapsed)
   - Left panel shows destroyed interference pattern (looks like single-slit)
   - Right panel shows measurements where first bit is 0 or 1 (which slit info)

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt

def enhanced_double_slit_simulation(num_slits=2,
                                   measure_slit=False,
                                   num_shots=40000):
    """
    - If measure_slit==False, you get true double-slit fringes.
    - If measure_slit==True, we collapse before any cry/H gates → single-slit.
    """
    qc = QuantumCircuit(4,4)

    # 1) If it's a true double-slit, put qubit 0 into |+⟩
    if num_slits == 2:
        qc.h(0)                    # <<< move H out of the measure_slit guard

    # 2) If you want which-path info, collapse qubit 0 immediately
    if measure_slit and num_slits == 2:
        qc.barrier()
        qc.measure(0, 0)
        qc.barrier()
        # now screen qubits will only see a definite |0> or |1> path

    # 3) Build interference only when you haven't peeked
    if not measure_slit and num_slits == 2:
        for q in [1,2,3]:
            theta = np.pi/2 * (q/3)
            qc.cry(theta, 0, q)
            qc.h(q)
    else:
        # single-slit / collapsed path → no two-path interference
        for q in [1,2,3]:
            qc.h(q)            # or skip entirely for a single-peak

    # 4) Final screen readout
    qc.barrier()
    qc.measure([1,2,3], [1,2,3])
    if not measure_slit:
        qc.measure(0,0)          # dummy-record the slit so full_counts has 4 bits

    # 5) Run and tally
    sim    = Aer.get_backend('qasm_simulator')
    job    = execute(qc, sim, shots=num_shots)
    counts = job.result().get_counts()

    # split out the dictionaries as before…
    full_counts   = counts
    screen_counts = {}
    slit_counts   = {'0':0, '1':0}

    for bits, c in counts.items():
        slit = bits[-1]    # now bit0 is in the *last* position
        screen = bits[:3]
        slit_counts[slit] += c
        screen_counts[screen] = screen_counts.get(screen, 0) + c

    return full_counts, screen_counts, slit_counts, qc

# Run simulations
full_double, screen_double, slit_double, qc1 = enhanced_double_slit_simulation(num_slits=2, measure_slit=False)
full_single, screen_single, slit_single, qc2 = enhanced_double_slit_simulation(num_slits=1, measure_slit=False)
full_measured, screen_measured, slit_measured, qc3 = enhanced_double_slit_simulation(num_slits=2, measure_slit=True)

#Display respective circuits
print("Double_slit:")
print(qc1.draw())

print("Single_slit:")
print(qc2.draw())

print("Which-path:")
print(qc3.draw())

# Create figure with 3 subplots
plt.figure(figsize=(18, 6))

# Subplot 1: Screen detection pattern
plt.subplot(1, 3, 1)
positions = sorted(screen_double.keys())
x = range(len(positions))

plt.bar(x, [screen_double[p] for p in positions], width=0.3, label='Double-slit')
plt.bar([i + 0.3 for i in x], [screen_single[p] for p in positions], width=0.3, label='Single-slit')
plt.bar([i + 0.6 for i in x], [screen_measured[p] for p in positions], width=0.3, label='Double-slit (measured)')

plt.xticks([i + 0.3 for i in x], positions)
plt.xlabel('Screen Position')
plt.ylabel('Counts')
plt.title('Interference Pattern at Detection Screen')
plt.legend()

# Subplot 2: Slit measurement results (when measured)
plt.subplot(1, 3, 2)
if sum(slit_measured.values()) > 0:
    plt.bar(['Left (0)', 'Right (1)'], [slit_measured['0'], slit_measured['1']], color=['blue', 'orange'])
    plt.title('Which-Slit Measurement Results')
    plt.ylabel('Counts')
else:
    plt.text(0.5, 0.5, 'No slit measurements taken\nin this simulation', 
             ha='center', va='center')
    plt.title('Which-Slit Measurement')

# Subplot 3: Full measurement outcomes (example)
plt.subplot(1, 3, 3)
example_counts = dict(list(full_double.items())[:8])  # Show first 8 outcomes
plt.bar(example_counts.keys(), example_counts.values())
plt.xticks(rotation=45)
plt.title('Example Full Measurements\n(slit + screen bits)')
plt.ylabel('Counts')

plt.tight_layout()
plt.show()
```

# More “screen” qubits for a finer-grained pattern

```python
from qiskit import QuantumCircuit, Aer, execute
import numpy as np
import matplotlib.pyplot as plt

def double_slit_simulation(num_slits=2,
                           num_screen_qubits=3,
                           which_path_measurement=False,
                           num_shots=20000):
    """
    Simulates a quantum double-slit with a variable number of 'screen' qubits.
      - num_slits: 1 or 2
      - num_screen_qubits: how many qubits to use for your 'screen'
      - which_path_measurement: True to collapse the slit-qubit before interference
      - num_shots: repetition count
    Returns:
      pattern_counts: dict mapping bitstrings to counts
      qc: the built circuit
    """
    # total qubits = 1 slit-qubit + N screen-qubits
    total_qubits = 1 + num_screen_qubits
    # classical bits = same as screen qubits (we ignore the slit bit unless measured)
    qc = QuantumCircuit(total_qubits, num_screen_qubits)

    # 1) prepare the “slit” qubit in superposition if double-slit
    if num_slits == 2:
        qc.h(0)

    # 2) build interference onto each screen-qubit
    #    we spread out phase shifts so the fringes have different spacings
    for j in range(num_screen_qubits):
        target = 1 + j
        # scale θ so the outer qubits see bigger phase-differences
        theta = np.pi/2 * ((j + 1) / num_screen_qubits)
        qc.cry(theta, 0, target)
        qc.h(target)

    qc.barrier()

    # 3) optionally collapse—i.e. “peek” at the path
    if which_path_measurement and num_slits == 2:
        # measure slit qubit into a scratch classical bit (we ignore it later)
        qc.measure(0, 0)

    # 4) measure all screen qubits into your classical register
    qc.measure(range(1, total_qubits), range(num_screen_qubits))

    # 5) run on qasm_simulator
    backend = Aer.get_backend('qasm_simulator')
    job     = execute(qc, backend, shots=num_shots)
    counts  = job.result().get_counts()

    # 6) massage counts into full bitstring → count dict
    #    (simulate missing leading-zero padding)
    pattern_counts = {
        format(i, f'0{num_screen_qubits}b'): 0
        for i in range(2**num_screen_qubits)
    }
    for bits, c in counts.items():
        screen_bits = bits[-num_screen_qubits:]
        pattern_counts[screen_bits] += c

    return pattern_counts, qc

# — example usage with num_screen_qubits for ultra-fine fringes —
num_screen_qubits=7
pattern5, qc5 = double_slit_simulation(
    num_slits=2,
    num_screen_qubits=num_screen_qubits,
    which_path_measurement=False,
    num_shots=20000
)

print(f"Circuit with {num_screen_qubits} screen qubits:")
print(qc5.draw())

# Plotting
positions = list(pattern5.keys())
counts    = [pattern5[p] for p in positions]

plt.figure(figsize=(10,4))
plt.bar(range(len(positions)), counts, width=0.8)
plt.xticks(range(len(positions)), positions, rotation=45, fontsize=4)
plt.title(f"Double-Slit Interference with {num_screen_qubits} Screen Qubits")
plt.xlabel(f"Screen Position ({num_screen_qubits}-bit string)")
plt.ylabel("Counts")
plt.tight_layout()
plt.show()
```
